#### Data Analysis

Let's start with a simple data analysis example. We'll use the pandas library to create a DataFrame and analyze it.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'Score': [85, 92, 88, 79]
}

df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:")
print(df)

# Calculate the average age
average_age = df['Age'].mean()
print("\nAverage Age:", average_age)

# Find the maximum score
max_score = df['Score'].max()
print("Maximum Score:", max_score)

# Visualize the data
plt.figure(figsize=(10, 5))

# Bar chart for scores
plt.subplot(1, 2, 1)
plt.bar(df['Name'], df['Score'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Score')
plt.title('Scores of Individuals')

# Scatter plot for age vs score
plt.subplot(1, 2, 2)
plt.scatter(df['Age'], df['Score'], color='green')
for i in range(len(df)):
    plt.text(df['Age'][i], df['Score'][i], df['Name'][i])
plt.xlabel('Age')
plt.ylabel('Score')
plt.title('Age vs Score')

plt.tight_layout()
plt.show()
```

#### User Interface with Input

Next, let's create a simple user interface that takes user input and displays a greeting message. We'll use the `ipywidgets` library for this purpose.

```python
%pip install -q ipywidgets
```

```python
import ipywidgets as widgets
from IPython.display import display

# Create a text input widget
name_input = widgets.Text(
    value='',
    placeholder='Type your name',
    description='Name:',
    disabled=False
)

# Create a button widget
button = widgets.Button(
    description='Greet',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click me',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)

# Define the callback function for the button
def on_button_click(b):
    print(f"Hello, {name_input.value}!")

# Attach the callback function to the button
button.on_click(on_button_click)

# Display the widgets
display(name_input)
display(button)
```
